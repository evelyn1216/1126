```
# encoding=utf-8
import jieba

from bs4 import BeautifulSoup

import urllib.request
contents = urllib.request.urlopen("https://www.thepaper.cn/").read()

soup = BeautifulSoup(contents, 'html.parser')

for t in soup.findAll("div", {"class": "news_li"}):

  url="https://www.thepaper.cn/"+t.a["href"]
  print(url)
  data = urllib.request.urlopen(url).read()

  bowl = BeautifulSoup(data, 'html.parser')
  for item in bowl.findAll("div", {"class": "news_txt"}):
    print(item.get_text())
    seg_list = jieba.cut(item.get_text(), cut_all=True)
    print("Full Mode: " + "/ ".join(seg_list))  # 全模式
```
